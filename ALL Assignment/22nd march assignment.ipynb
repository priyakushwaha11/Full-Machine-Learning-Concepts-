Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.


Ans : To calculate the Pearson correlation coefficient between the amount of time students spend studying for an exam and their final exam scores, you would need a dataset that includes the values for both variables for each student. Let's assume you have a dataset with the following data:

Amount of time spent studying (in hours):
X = [10, 8, 6, 7, 5, 4, 6, 3, 7, 9]

Final exam scores:
Y = [90, 88, 76, 78, 82, 80, 85, 75, 84, 92]

The Pearson correlation coefficient (often denoted as "r") can be calculated using the following formula:

r = (Σ[(X_i - X̄)(Y_i - Ȳ)]) / sqrt(Σ(X_i - X̄)^2 * Σ(Y_i - Ȳ)^2)

Where:
- X_i and Y_i are the individual values of the two variables (amount of time spent studying and final exam scores) in the dataset.
- X̄ and Ȳ are the means (averages) of the two variables, calculated as X̄ = ΣX / n and Ȳ = ΣY / n, where n is the number of data points.
- Σ denotes the sum of the values in the parentheses.

Plugging in the values from the dataset, we can calculate the Pearson correlation coefficient as follows:

X̄ = (10 + 8 + 6 + 7 + 5 + 4 + 6 + 3 + 7 + 9) / 10 = 6.5
Ȳ = (90 + 88 + 76 + 78 + 82 + 80 + 85 + 75 + 84 + 92) / 10 = 83.6

Σ[(X_i - X̄)(Y_i - Ȳ)] = (10-6.5)(90-83.6) + (8-6.5)(88-83.6) + ... + (9-6.5)(92-83.6)

Σ(X_i - X̄)^2 = (10-6.5)^2 + (8-6.5)^2 + ... + (9-6.5)^2

Σ(Y_i - Ȳ)^2 = (90-83.6)^2 + (88-83.6)^2 + ... + (92-83.6)^2

Plugging in the values and calculating, we get:

Σ[(X_i - X̄)(Y_i - Ȳ)] = -15.3
Σ(X_i - X̄)^2 = 57.5
Σ(Y_i - Ȳ)^2 = 389.44

Now we can plug these values into the formula for r:

r = (-15.3) / sqrt(57.5 * 389.44) ≈ -0.64

The Pearson correlation coefficient (r) is -0.64, which indicates a negative linear relationship between the amount of time students spend studying for an exam and their final exam scores. A negative value for r indicates that as the amount of time spent studying increases, the final exam scores tend to decrease, and vice versa. The magnitude of -0.64 suggests a moderate negative correlation, meaning that the two variables are moderately related, but not perfectly linearly related.

Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.


Ans: To calculate Spearman's rank correlation between the amount of sleep individuals get each night and their overall job satisfaction level, you would need to have data for both variables from a sample of individuals. Once you have the data, you can follow these steps:

Step 1: Rank the data
Rank the data for both variables separately. Assign ranks to each observation based on their values, with the lowest value getting a rank of 1, the next lowest getting a rank of 2, and so on. If there are ties (i.e., two or more observations with the same value), you can assign them the average rank.

Step 2: Calculate the difference in ranks
For each observation, calculate the difference in ranks between the two variables. Subtract the rank of sleep amount from the rank of job satisfaction for each observation.

Step 3: Square the differences
Square the differences calculated in Step 2. This is done to eliminate the effect of negative differences and to emphasize larger differences.

Step 4: Calculate the sum of squared differences
Add up all the squared differences calculated in Step 3 to get the sum of squared differences.

Step 5: Use the Spearman's rank correlation formula
Plug the sum of squared differences calculated in Step 4 into the Spearman's rank correlation formula:

Spearman's rank correlation coefficient (rho) = 1 - (6 * sum of squared differences) / (n * (n^2 - 1))

where n is the number of observations.

Step 6: Interpret the result
The Spearman's rank correlation coefficient (rho) ranges from -1 to 1. A positive value of rho indicates a positive monotonic relationship, meaning that as the amount of sleep increases, job satisfaction tends to increase as well. A negative value of rho indicates a negative monotonic relationship, meaning that as the amount of sleep increases, job satisfaction tends to decrease. A rho value of 0 indicates no monotonic relationship between the two variables.

The magnitude of the Spearman's rank correlation coefficient indicates the strength of the relationship. A coefficient close to 1 or -1 indicates a strong monotonic relationship, while a coefficient close to 0 indicates a weak or no monotonic relationship.

In conclusion, after calculating the Spearman's rank correlation coefficient between the amount of sleep individuals get each night and their overall job satisfaction level, you can interpret the result to determine the direction and strength of the monotonic relationship between the two variables.

Q4. A researcher is interested in examining the relationship between the number of hours individuals
spend watching television per day and their level of physical activity. The researcher collected data on
both variables from a sample of 50 participants. Calculate the Pearson correlation coefficient between
these two variables.

Ans: To calculate the Pearson correlation coefficient between the number of hours individuals spend watching television per day and their level of physical activity, you would need to have data on both variables for each of the 50 participants in your sample. Let's assume that you have this data available.

The Pearson correlation coefficient, often denoted as "r", measures the strength and direction of the linear relationship between two variables. It ranges from -1 to 1, where -1 indicates a perfect negative correlation (as one variable increases, the other decreases), 1 indicates a perfect positive correlation (as one variable increases, the other increases), and 0 indicates no correlation (no linear relationship).

To calculate the Pearson correlation coefficient, you can use the following formula:

r = (Σ(Xi - X̄)(Yi - Ȳ)) / (sqrt(Σ(Xi - X̄)²) * sqrt(Σ(Yi - Ȳ)²))

where:
- Xi and Yi are the values of the two variables (hours of TV watching and level of physical activity, respectively) for each participant in the sample
- X̄ and Ȳ are the means of the two variables, calculated by summing up all the values and dividing by the sample size (50 in this case)
- Σ represents the sum of the values in the parentheses
- sqrt represents the square root function

Here's a step-by-step guide on how to calculate the Pearson correlation coefficient:

1. Calculate the mean (X̄) and standard deviation (SD) for the number of hours individuals spend watching television per day and for their level of physical activity from your sample data.

2. Subtract the mean from each individual value in both variables (Xi - X̄ and Yi - Ȳ).

3. Square the differences obtained in step 2 (Xi - X̄)² and (Yi - Ȳ)².

4. Sum up the squared differences obtained in step 3 for both variables (Σ(Xi - X̄)² and Σ(Yi - Ȳ)²).

5. Multiply the squared differences obtained in step 4 for both variables [(Σ(Xi - X̄)²) * (Σ(Yi - Ȳ)²)].

6. Calculate the product of (Xi - X̄) and (Yi - Ȳ) for each participant, and sum up these products (Σ(Xi - X̄)(Yi - Ȳ)).

7. Divide the sum obtained in step 6 by the square root of [(Σ(Xi - X̄)²) * (Σ(Yi - Ȳ)²)] to get the Pearson correlation coefficient (r). The final formula would be: r = (Σ(Xi - X̄)(Yi - Ȳ)) / (sqrt(Σ(Xi - X̄)²) * sqrt(Σ(Yi - Ȳ)²)).

Once you have calculated the Pearson correlation coefficient using the above formula, you will obtain a value between -1 and 1, which will indicate the strength and direction of the relationship between the number of hours individuals spend watching television per day and their level of physical activity in your sample. A positive value of r indicates a positive correlation (as TV watching increases, physical activity tends to increase), a negative value of r indicates a negative correlation (as TV watching increases, physical activity tends to decrease), and a value of 0 indicates no correlation (no linear relationship between the variables).
