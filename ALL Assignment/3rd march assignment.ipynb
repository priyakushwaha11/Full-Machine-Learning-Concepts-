Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans : 

1. **Scatter Plot**: Scatter plot is used to display the relationship between two continuous variables. It is commonly used in data visualization to identify patterns and trends in the data. It is plotted using the scatterplot() function in Seaborn.


2. **Line Plot**: Line plot is used to display the trend of a continuous variable over time or any other continuous variable. It is plotted using the lineplot() function in Seaborn.


3. **Bar Plot**: Bar plot is used to compare different categories or groups of data. It can be used to visualize the frequency distribution of a categorical variable. It is plotted using the barplot() function in Seaborn.


4. **Histogram**: Histogram is used to display the frequency distribution of a continuous variable. It divides the data into intervals and displays the frequency of each interval using bars. It is plotted using the histplot() function in Seaborn.


5. **Box Plot**: Box plot is used to display the distribution of a continuous variable. It shows the median, quartiles, and outliers of the data. It is plotted using the boxplot() function in Seaborn. 



Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

Note: timepoint, signal, event, and region are columns in the fmri dataset.

**Ans** :

```python
import seaborn as sns

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Plot line plot for different events and regions
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

Note: pclass, age, and fare are columns in the titanic dataset.

**Ans**:

```python
import seaborn as sns

# Load titanic dataset
titanic = sns.load_dataset("titanic")

# Create two box plots using seaborn
sns.boxplot(x='pclass', y='age', data=titanic)
sns.boxplot(x='pclass', y='fare', data=titanic)
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

**Ans**:

```python
import seaborn as sns

diamonds = sns.load_dataset('diamonds')
```

```python
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack')
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

**Ans**:

```python
import seaborn as sns

# load the iris dataset
iris = sns.load_dataset("iris")

# use the pairplot function with hue='species'
sns.pairplot(iris, hue='species')
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

**Ans**

```python
import seaborn as sns

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Plot a heatmap using the pivot table of the data
sns.heatmap(flights_data.pivot("month", "year", "passengers"), annot=True, fmt="d")
```
