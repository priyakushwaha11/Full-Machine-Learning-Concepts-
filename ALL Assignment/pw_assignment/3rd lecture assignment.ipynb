1. Scenario: A company wants to analyze the sales performance of its products in different regions. They have collected the following data:
   Region A: [10, 15, 12, 8, 14]
   Region B: [18, 20, 16, 22, 25]
   Calculate the mean sales for each region.


ANS: To calculate the mean sales for each region, you need to add up all the sales values in each region and then divide the sum by the number of data points. Let's calculate the mean sales for Region A and Region B based on the given data:


Region A: [10, 15, 12, 8, 14]


To find the mean sales for Region A, we sum up all the values and divide by the number of data points:


Mean for Region A = (10 + 15 + 12 + 8 + 14) / 5
= 59 / 5
= 11.8


Therefore, the mean sales for Region A is 11.8.


Region B: [18, 20, 16, 22, 25]


To find the mean sales for Region B, we sum up all the values and divide by the number of data points:


Mean for Region B = (18 + 20 + 16 + 22 + 25) / 5
= 101 / 5
= 20.2


Therefore, the mean sales for Region B is 20.2.


So, the mean sales for Region A is 11.8 and for Region B is 20.2.

4. Scenario: A data analyst wants to determine the variability in the daily stock prices of a company. The data collected is as follows:
   [25.5, 24.8, 26.1, 25.3, 24.9]
   Calculate the range of the stock prices.


ANS: To calculate the range of the stock prices, you need to find the difference between the highest and lowest values in the given data set. In this case, the data set is [25.5, 24.8, 26.1, 25.3, 24.9].


First, let's arrange the stock prices in ascending order:
[24.8, 24.9, 25.3, 25.5, 26.1]


The lowest value is 24.8, and the highest value is 26.1.


Range = Highest value - Lowest value
= 26.1 - 24.8
= 1.3


Therefore, the range of the stock prices is 1.3.

5. Scenario: A study is conducted to compare the performance of two different teaching methods. The test scores of the students in each group are as follows:
   Group A: [85, 90, 92, 88, 91]
   Group B: [82, 88, 90, 86, 87]
   Perform a t-test to determine if there is a significant difference in the mean scores between the two groups.


ANS: To perform a t-test to determine if there is a significant difference in the mean scores between two groups, we need to compare the means of the two groups and calculate the t-statistic and p-value. The t-test assumes that the data is normally distributed and the variances of the two groups are equal.


Let's perform the t-test using the given data:


Group A: [85, 90, 92, 88, 91]
Group B: [82, 88, 90, 86, 87]


Step 1: Calculate the means of both groups.
Mean of Group A = (85 + 90 + 92 + 88 + 91) / 5 = 88.4
Mean of Group B = (82 + 88 + 90 + 86 + 87) / 5 = 86.6


Step 2: Calculate the variances of both groups.
Variance of Group A = ((85 - 88.4)^2 + (90 - 88.4)^2 + (92 - 88.4)^2 + (88 - 88.4)^2 + (91 - 88.4)^2) / 4 = 4.3
Variance of Group B = ((82 - 86.6)^2 + (88 - 86.6)^2 + (90 - 86.6)^2 + (86 - 86.6)^2 + (87 - 86.6)^2) / 4 = 3.2


Step 3: Calculate the pooled standard deviation.
Pooled Standard Deviation = sqrt(((n1 - 1) * Var1 + (n2 - 1) * Var2) / (n1 + n2 - 2))
= sqrt(((5 - 1) * 4.3 + (5 - 1) * 3.2) / (5 + 5 - 2))
= sqrt((16.8 + 12.8) / 8)
= sqrt(29.6 / 8)
= sqrt(3.7)
= 1.92


Step 4: Calculate the t-statistic.
t-statistic = (Mean1 - Mean2) / (Pooled Standard Deviation * sqrt(1/n1 + 1/n2))
= (88.4 - 86.6) / (1.92 * sqrt(1/5 + 1/5))
= 1.8 / (1.92 * sqrt(0.2 + 0.2))
= 1.8 / (1.92 * sqrt(0.4))
= 1.8 / (1.92 * 0.632)
= 1.8 / 1.216
= 1.48


Step 5: Determine the degrees of freedom (df).
Degrees of Freedom = n1 + n2 - 2 = 5 + 5 - 2 = 8


Step 6: Determine the critical t-value for the desired significance level (α).
Assuming a significance level (α) of 0.05, a two-tailed t-test with 8 degrees of freedom has a critical t-value of approximately ±2.31.


Step 7: Compare the t-statistic with the critical t-value and calculate the p-value.
As the absolute value of the calculated t-statistic (1.48) is less than the critical t-value (2.31), we fail to reject the null hypothesis.


Therefore, based on the t-test, there is no significant difference in the mean scores between Group A and Group B.


6. Scenario: A company wants to analyze the relationship between advertising expenditure and sales. The data collected is as follows:
   Advertising Expenditure (in thousands): [10, 15, 12, 8, 14]
   Sales (in thousands): [25, 30, 28, 20, 26]
   Calculate the correlation coefficient between advertising expenditure and sales.


ANS: To calculate the correlation coefficient between advertising expenditure and sales, we can use the Pearson correlation coefficient formula. The Pearson correlation coefficient measures the linear relationship between two variables.


Let's calculate the correlation coefficient using the given data:


Advertising Expenditure (in thousands): [10, 15, 12, 8, 14]
Sales (in thousands): [25, 30, 28, 20, 26]


Step 1: Calculate the means of both variables.
Mean of Advertising Expenditure = (10 + 15 + 12 + 8 + 14) / 5 = 11.8
Mean of Sales = (25 + 30 + 28 + 20 + 26) / 5 = 25.8


Step 2: Calculate the deviations from the means for both variables.
Deviation from the mean of Advertising Expenditure: [-1.8, 3.2, 0.2, -3.8, 2.2]
Deviation from the mean of Sales: [-0.8, 4.2, 2.2, -5.8, 0.2]


Step 3: Calculate the product of the deviations for each pair of data points.
Product of deviations: [(-1.8 * -0.8), (3.2 * 4.2), (0.2 * 2.2), (-3.8 * -5.8), (2.2 * 0.2)] = [1.44, 13.44, 0.44, 22.04, 0.44]


Step 4: Calculate the sum of the products of deviations.
Sum of products of deviations = 1.44 + 13.44 + 0.44 + 22.04 + 0.44 = 37.8


Step 5: Calculate the standard deviation of both variables.
Standard deviation of Advertising Expenditure = sqrt(((-1.8)^2 + 3.2^2 + 0.2^2 + (-3.8)^2 + 2.2^2) / 4) = sqrt(29.04 / 4) = sqrt(7.26) = 2.69
Standard deviation of Sales = sqrt(((-0.8)^2 + 4.2^2 + 2.2^2 + (-5.8)^2 + 0.2^2) / 4) = sqrt(62.76 / 4) = sqrt(15.69) = 3.96


Step 6: Calculate the correlation coefficient.
Correlation coefficient = (Sum of products of deviations) / (Standard deviation of Advertising Expenditure * Standard deviation of Sales)
= 37.8 / (2.69 * 3.96)
= 37.8 / 10.64
= 3.55


Therefore, the correlation coefficient between advertising expenditure and sales is approximately 3.55. The value indicates a strong positive correlation between the two variables.
