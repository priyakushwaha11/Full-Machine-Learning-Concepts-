```python
Q1. What is Lasso Regression, and how does it differ from other regression techniques?

Ans: Lasso Regression, or L1 regularization, is a linear regression technique that adds a penalty term based on the 
    absolute values of the coefficients to the ordinary least squares (OLS) objective function. This penalty term helps
    in regularizing the model by shrinking some coefficients to exactly zero, effectively performing feature selection.
    The main difference from other regression techniques, such as Ridge Regression, is the type of penalty applied: 
        Lasso uses L1 penalty, while Ridge uses L2 penalty.
        
        
Q2. What is the main advantage of using Lasso Regression in feature selection?

Ans: The main advantage of using Lasso Regression in feature selection is that it tends to produce sparse models by 
    setting some coefficients to exactly zero. This means that Lasso can automatically select a subset of relevant
    features, leading to a more interpretable and efficient model.
    
    
Q3. How do you interpret the coefficients of a Lasso Regression model?

Ans:Q3. The coefficients of a Lasso Regression model can be interpreted in terms of feature importance. A non-zero
    coefficient indicates the importance of the corresponding feature in predicting the target variable. Coefficients
    that are exactly zero imply that the associated features have been excluded from the model. 
    
    
Q4. What are the tuning parameters that can be adjusted in Lasso Regression, and how do they affect the
model's performance?

Ans:  In Lasso Regression, the tuning parameter is usually denoted as lambda (Î»). It controls the strength of the
    regularization penalty. As lambda increases, more coefficients are pushed towards zero, leading to increased 
    sparsity in the model. The tuning parameter can be adjusted to find the right balance between model simplicity and
    predictive performance.
    
    
Q5. Can Lasso Regression be used for non-linear regression problems? If yes, how?

Ans: Lasso Regression is inherently a linear regression technique and may not handle non-linear relationships between
    features and the target variable well. However, it can be used for non-linear regression problems by incorporating
    non-linear transformations of the input features or by combining it with non-linear modeling techniques.
    
    
Q6. What is the difference between Ridge Regression and Lasso Regression?

Ans: The main difference between Ridge Regression and Lasso Regression lies in the type of penalty applied. Ridge 
    Regression uses an L2 penalty, which penalizes the squared values of the coefficients, while Lasso uses an 
    L1 penalty, penalizing the absolute values of the coefficients. The L1 penalty tends to produce sparse models
    with some coefficients exactly zero, making Lasso useful for feature selection.
    
    
Q7. Can Lasso Regression handle multicollinearity in the input features? If yes, how?

Ans: Yes, Lasso Regression can handle multicollinearity to some extent. The L1 penalty tends to select one feature 
    among a group of highly correlated features, effectively ignoring the others. This can be advantageous in 
    situations where multicollinearity might be present, as it automatically chooses a subset of features.
    
    
Q8. How do you choose the optimal value of the regularization parameter (lambda) in Lasso Regression?

Ans: The optimal value of the regularization parameter (lambda) in Lasso Regression is typically chosen through
    cross-validation. By training the model on different subsets of the data and evaluating its performance, one can 
    identify the lambda value that provides the best balance between model simplicity and predictive accuracy. 
    Cross-validation techniques, such as k-fold cross-validation, are commonly used for this purpose.
```
