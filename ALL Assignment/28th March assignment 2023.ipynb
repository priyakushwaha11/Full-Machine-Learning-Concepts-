```python
Q1. What is Ridge Regression, and how does it differ from ordinary least squares regression?

Ans:Ridge Regression: Ridge Regression is a linear regression technique that includes a regularization term
        (L2 regularization) in the cost function. The regularization term is the sum of squared magnitudes of 
        coefficients multiplied by a tuning parameter (lambda). The objective is to penalize large coefficients, 
        preventing overfitting and improving the model's generalization performance.
    
    
Q2. What are the assumptions of Ridge Regression?

Ans:Ridge Regression assumes a linear relationship between the independent and dependent variables.
It assumes that the residuals (the differences between observed and predicted values) are normally distributed.
Assumption of homoscedasticity: The variance of the residuals is constant across all levels of the independent variables.
Independence of residuals: The residuals should be independent of each other.  
    
    
Q3. How do you select the value of the tuning parameter (lambda) in Ridge Regression?

Ans: The value of lambda is often chosen through techniques like cross-validation. The goal is to select the value that
    minimizes prediction error while also preventing overfitting.
       
    
Q4. Can Ridge Regression be used for feature selection? If yes, how?

Ans:Yes, Ridge Regression can shrink the coefficients of less important features towards zero. While it doesn't 
    exactly perform feature selection (i.e., setting some coefficients exactly to zero), it effectively downweights
    less important features. 
    
    
Q5. How does the Ridge Regression model perform in the presence of multicollinearity?

Ans: Ridge Regression is effective in handling multicollinearity. It introduces a penalty term that prevents the model 
    from relying too heavily on any single variable, reducing the impact of multicollinearity on coefficient estimates.
    
    
    
Q6. Can Ridge Regression handle both categorical and continuous independent variables?

Ans: Yes, Ridge Regression can handle both categorical and continuous independent 
    variables. For categorical variables, they are typically converted into dummy variables before applying Ridge 
    Regression.
    
    
    
    
Q7. How do you interpret the coefficients of Ridge Regression?

Ans: 
    
    
    
    
Q8. Can Ridge Regression be used for time-series data analysis? If yes, how?

Ans: 
    
    
    
```
