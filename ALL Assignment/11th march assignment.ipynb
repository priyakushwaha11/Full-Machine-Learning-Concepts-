Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would
use each type of test.

ANS: 

A t-test and a z-test are both statistical tests used to make inferences about population parameters based on sample data. However, they differ in their assumptions and the type of data they are used for.

A z-test is used when the population standard deviation is known, or when the sample size is sufficiently large (typically n ≥ 30) that the sample standard deviation can be assumed to be a good estimate of the population standard deviation. It is appropriate for continuous data that are normally distributed or approximately so. 

For example, suppose you want to test whether the average height of male college students in the US is different from 6 feet. You could take a sample of 50 male college students and measure their heights, and then conduct a z-test. If the resulting test statistic falls outside a critical value based on a chosen level of significance (such as p < 0.05), you would reject the null hypothesis that the average height is 6 feet and conclude that the true population mean is likely different from 6 feet.

On the other hand, a t-test is used when the population standard deviation is unknown and must be estimated from the sample data, or when the sample size is small (typically n < 30). It is appropriate for continuous data that are normally distributed or approximately so. 

For example, suppose you want to test whether a new drug has a different effect on blood pressure compared to a placebo. You could randomly assign 20 patients to either the treatment or control group, measure their blood pressure before and after the intervention, and conduct a t-test. If the resulting test statistic falls outside a critical value based on a chosen level of significance (such as p < 0.05), you would reject the null hypothesis that the drug has no effect on blood pressure and conclude that the drug has a different effect than the placebo.

In summary, a z-test is used when the population standard deviation is known or the sample size is large, while a t-test is used when the population standard deviation is unknown or the sample size is small.

Q2: Differentiate between one-tailed and two-tailed tests.

ANS: 

In statistics, a hypothesis test is a way of making decisions based on data about whether a particular hypothesis is true or not. The null hypothesis is the statement that is being tested, while the alternative hypothesis is the opposite of the null hypothesis. There are two types of hypothesis tests: one-tailed tests and two-tailed tests.

A one-tailed test is a hypothesis test where the alternative hypothesis specifies the direction of the effect. That is, the alternative hypothesis predicts that the observed effect will be either greater than or less than a certain value, but not both. For example, a one-tailed test could be used to determine whether a new drug reduces blood pressure (one-tailed test for a decrease) or whether a new marketing campaign increases sales (one-tailed test for an increase).

A two-tailed test is a hypothesis test where the alternative hypothesis does not specify the direction of the effect. That is, the alternative hypothesis predicts that the observed effect will be different from a certain value, either greater than or less than that value. For example, a two-tailed test could be used to determine whether a new drug affects blood pressure (two-tailed test for any change) or whether a new marketing campaign affects sales (two-tailed test for any change).

The choice of a one-tailed or two-tailed test depends on the specific research question and the nature of the effect being studied. A one-tailed test is more powerful than a two-tailed test, meaning that it is more likely to detect a true effect if one exists. However, it is also more prone to type I errors (false positives) if the direction of the effect is not correctly specified. A two-tailed test is more conservative and less prone to type I errors, but it is also less powerful than a one-tailed test.

Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for
each type of error.

ANS: 

Type 1 and Type 2 errors are two types of errors that can occur in statistical hypothesis testing. 

Type 1 error, also known as a false positive, occurs when we reject a null hypothesis that is actually true. In other words, we conclude that there is a significant difference between two groups when in reality, there is no such difference. The probability of making a Type 1 error is denoted by alpha (α) and is usually set at 0.05 (5%).

An example of a Type 1 error would be a medical test that shows a patient has a disease when they actually don't. This can lead to unnecessary treatments and interventions, which can be harmful to the patient.

Type 2 error, also known as a false negative, occurs when we fail to reject a null hypothesis that is actually false. In other words, we conclude that there is no significant difference between two groups when in reality, there is a difference. The probability of making a Type 2 error is denoted by beta (β).

An example of a Type 2 error would be a medical test that shows a patient does not have a disease when they actually do. This can lead to delayed or missed diagnoses, which can have serious consequences for the patient's health.

In hypothesis testing, the goal is to minimize both Type 1 and Type 2 errors. However, there is often a trade-off between the two. By lowering the probability of one type of error, we increase the probability of the other. This is known as the "power" of the test and is a function of the sample size, effect size, and significance level.

Q4: Explain Bayes's theorem with an example.

ANS: 

Bayes's theorem is a fundamental concept in probability theory that describes the relationship between conditional probabilities. It provides a way to calculate the probability of an event based on prior knowledge or information.

The theorem is named after the English statistician Thomas Bayes, who first formulated it in the 18th century. It can be stated as follows:

P(A|B) = P(B|A) * P(A) / P(B)

Where P(A|B) is the probability of event A occurring given that event B has occurred, P(B|A) is the probability of event B occurring given that event A has occurred, P(A) is the prior probability of event A occurring, and P(B) is the prior probability of event B occurring.

To illustrate the theorem, let's consider an example. Suppose we have a test for a rare disease, which affects only 1% of the population. The test has a 95% accuracy rate, meaning that it correctly identifies 95% of people with the disease and 5% of people without the disease will be false positives. 

Now suppose a person tests positive for the disease. What is the probability that they actually have the disease?

Using Bayes's theorem, we can calculate the probability as follows:

P(Disease|Positive) = P(Positive|Disease) * P(Disease) / P(Positive)

We know that P(Disease) = 0.01 (the prior probability of having the disease), and P(Positive|Disease) = 0.95 (the probability of testing positive given that the person has the disease). The probability of testing positive, P(Positive), can be calculated as:

P(Positive) = P(Positive|Disease) * P(Disease) + P(Positive|No Disease) * P(No Disease)
             = 0.95 * 0.01 + 0.05 * 0.99
             = 0.058

So, the probability of actually having the disease given a positive test result is:

P(Disease|Positive) = 0.95 * 0.01 / 0.058
                     = 0.1638
                     = 16.38%

Therefore, even with a positive test result, the probability of having the disease is still relatively low (about 16.38%). This is because the prior probability of having the disease is very low and the false positive rate of the test is relatively high.
    

Q5: What is a confidence interval? How to calculate the confidence interval, explain with an example.

ANS:

A confidence interval is a statistical range or interval estimate that quantifies the uncertainty around a sample statistic such as the mean, standard deviation, proportion, or regression coefficient. It is a range of values within which we can be confident that the true population parameter lies, given a certain level of confidence. For example, if we say that a certain parameter has a 95% confidence interval of (0.4, 0.6), we mean that if we repeated the sampling process many times and constructed 95% confidence intervals from each sample, about 95% of those intervals would contain the true value of the parameter.

To calculate a confidence interval, we need to have a sample statistic, the sample size, and the desired level of confidence. The formula for calculating a confidence interval depends on the sample statistic and the sampling distribution. For example, if we have a sample of n observations from a normally distributed population with unknown mean μ and known standard deviation σ, we can use the following formula to construct a confidence interval for the population mean μ at a confidence level of 1-α:

x̄ ± z(α/2) * (σ/√n)

where x̄ is the sample mean, z(α/2) is the z-score corresponding to the desired level of confidence (e.g., 1.96 for a 95% confidence level), σ is the population standard deviation, and n is the sample size.

For example, suppose we want to estimate the mean height of adult males in a certain population. We take a random sample of 50 adult males and measure their heights, obtaining a sample mean of 175 cm and a sample standard deviation of 8 cm. We want to construct a 95% confidence interval for the population mean height.

Using the formula above, we have:

175 ± 1.96 * (8/√50)

= (171.7, 178.3)

Therefore, we can be 95% confident that the true population mean height lies within the interval (171.7, 178.3) based on our sample of 50 adult males.
