Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.


Ans - 
Matplotlib is a popular data visualization library for Python that allows users to create a wide range of static, animated, and interactive visualizations in Python. It was initially developed in 2003 by John D. Hunter and has since become one of the most widely used libraries for scientific computing with Python.

Matplotlib is used for creating visualizations of various types of data such as time-series, scatter plots, histograms, bar plots, etc. It provides a wide range of customization options to make the visualizations look exactly the way the user wants.

Pyplot is a sub-module of Matplotlib that provides a MATLAB-like interface for creating visualizations. Some of the popular plots that can be created using the Pyplot module of Matplotlib are:

1. Line Plot - Used for showing trends or changes over time.

2. Scatter Plot - Used for showing the relationship between two variables.

3. Histogram - Used for showing the distribution of a single variable.

4. Bar Chart - Used for comparing multiple categories with a single metric.

5. Pie Chart - Used for showing the proportion of each category in a whole.

 Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

Ans. 

A scatter plot is a type of plot that displays values for two variables as points on a Cartesian plane, where one variable is plotted on the x-axis and the other on the y-axis. Scatter plots are useful for showing the relationship between two variables and identifying patterns or trends in the data.

Here is the code to generate a scatter plot using the given data:


```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter Plot of X and Y")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans. 

The subplot() function is used to create multiple plots in a single figure. It allows us to display multiple plots in a grid-like structure, where each plot represents a subset of the data. This function takes three arguments: the number of rows, the number of columns, and the plot number.

Here is an example code that generates four line plots using the subplot() function with the data provided:



```python
import numpy as np
import matplotlib.pyplot as plt

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Plot line 1 in the first subplot
axs[0, 0].plot(x1, y1)

# Plot line 2 in the second subplot
axs[0, 1].plot(x2, y2)

# Plot line 3 in the third subplot
axs[1, 0].plot(x3, y3)

# Plot line 4 in the fourth subplot
axs[1, 1].plot(x4, y4)

# Add a title to the figure
fig.suptitle('Four Line Plots')

# Show the plot
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

Ans . A bar plot is a type of data visualization that presents categorical data with rectangular bars, where the length of each bar corresponds to the value of the category it represents. Bar plots are commonly used to show comparisons between different categories, and they can be used to display either numerical or categorical data.

```python
import matplotlib.pyplot as plt
import numpy as np

# create the data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# create a bar plot
plt.bar(company, profit)
plt.title('Company Profit')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()

# create a horizontal bar plot
plt.barh(company, profit)
plt.title('Company Profit')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

Ans. A box plot is a graphical representation of the distribution of a dataset that displays the median, quartiles, and range of the data. The box in the plot represents the middle 50% of the data, with the bottom of the box indicating the first quartile (Q1) and the top of the box indicating the third quartile (Q3). The line inside the box indicates the median value, and the whiskers extending from the box show the minimum and maximum values of the data. Any data points outside of the whiskers are considered outliers and are plotted as individual points.

```python
import seaborn as sns
import numpy as np

# Generate the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data into a single array
data = np.concatenate([box1, box2])

# Create the box plot
sns.boxplot(data=data)
```
