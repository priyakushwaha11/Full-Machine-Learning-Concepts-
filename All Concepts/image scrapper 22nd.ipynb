```python
import requests 
from bs4 import BeautifulSoup 
from urllib.request import urlopen as uReq
import logging
import os
import pymongo
```

```python
save_dir = "image/"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
quary = "shahrukh khan"
response = requests.get (f"https://www.google.com/search?q{quary}=shahrukh+khan&source=lmns&bih=577&biw=1263&rlz=1C1RXQR_enIN1037IN1037&hl=en&sa=X&ved=2ahUKEwjYxYbB1bX9AhVo2XMBHbmlD7gQ_AUoAHoECAEQAA")
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content , 'html.parser')
```

```python
soup
```

```python
images_tag = soup.find_all("img")
```

```python
len(images_tag)
```

```python
#images ka length jyada hone per loop chalate hain ki ek ek image 
#directory me push karne ke liye 13no. wali line chalate hain 
#mongo db me karne ke liye 18no. wali line chalate hain
#wb ka full form web binary

img_data_mongo = []
for i in images_tag:
    image_url = i['src']
    image_data = requests.get(image_url).content
    mydict = {
        "index" : image_url,
        "image" : image_data
    }
    img_data_mongo.append(mydict)
    with open (os.path.join(save_dir,f"{quary}_{image_tag.index(i)}.jpg"),"wb") as f :
        f.write (image_data)
        

#client = pymongo.MongoClient("mongodb+srv://priyakushwaha9669:<priyakushwaha>@cluster0.y2epfli.mongodb.net/?retryWrites=true&w=majority")
#db = client.test['image_scrab']
#coll_image = db['image_scrab']
#coll_image.insert_many(img_data_mongo)
```
