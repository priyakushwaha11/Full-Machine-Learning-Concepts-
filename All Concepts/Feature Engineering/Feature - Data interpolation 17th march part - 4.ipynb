## linear interpolation

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5,6,7,8])
y = np.array([3.4,5.6,5.5,5.3,2,5.4,4.3,6.6])
```

```python
# interpolate the data using linear interpolate

import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
#  interpolate the data using linear interpolation

x_new = np.linspace(1,5,10)  ## create x value
y_interp = np.interp(x_new,x,y)  ## interpolate y value
```

```python
y_interp
```

```python
plt.scatter(x_new , y_interp)
```

### Cubic interpolation with scipy

```python
import numpy as np
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])
```

```python
x,y
```

```python
from scipy.interpolate import interp1d
```

```python
## create a cubic interpolation function

f = interp1d(x,y,kind='cubic')
```

```python
## interpolate the data

x_new = np.linspace(1,5,10)
y_interp = f(x_new)
print(y_interp)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```

### polynomial interpolation
