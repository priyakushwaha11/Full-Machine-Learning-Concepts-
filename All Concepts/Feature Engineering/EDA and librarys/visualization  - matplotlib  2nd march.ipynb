```python
# categorical/string and numerical data ke liye

x = ['a','b','c','d','e','f']
y = np.random.rand(6)
plt.bar(x,y)
```

```python
y
```

```python
plt.bar(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar plot")
```

```python
# Horizontal plot banane ke liye 

plt.barh(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar plot")
```

```python
plt.plot(x,y ,'--r')
plt.show
```

```python
plt.figure(figsize=(2,6))
```

```python
plt.figure(figsize=(2,6))
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar plot")
```

```python
plt.figure(figsize=(2,6))
plt.plot(x,y)
plt.scatter(x,y , marker='v')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("bar plot")
plt.grid ()
plt.show()
```
