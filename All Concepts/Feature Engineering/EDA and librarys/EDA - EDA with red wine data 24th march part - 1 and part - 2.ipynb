```python
# Only categorical features se numerical features alag karne ke liye

numerical_feature = [feature for feature in df.columns if df[feature].dtype!='O']
categorical_feature = [feature for feature in df.columns if df[feature].dtype=='O']
```

```python
numerical_feature
```

```python
categorical_feature
```

```python
# categorical data ki unique value nikalne ke liye

df['gender'].value_counts()
```

```python
df['race_ethnicity'].value_counts()
```

```python
#aggregate the total score with mean mtlb ek ek categorical data ke poore poore numerical wale column ko add karke average / total score nikalne ke liye

df['total_score'] = (df['math_score'] + df['reading_score'] + df['writing_score'])
df['average_score'] = df['total_score'] / 3
```
