```python
## Feature engineering

df.loc[:122,"Region"]=0
df.loc[122:,"Region"]=1
df1=df

# 122nd rows tk region me 0 show karega ushke aage 1 show hoga



# output me region add ho jayega 
```

#### Changes the required columns as integer data type

```python
# object ko int me change karne ke liye

df1[['day','month','year','Temperature','RH', 'Ws']] = df1[['day','month','year','Temperature','RH', 'Ws']].astype(int)
```

#### Changing the column float data type

```python
df1.head(4)
```

```python
## bache huye object poore features ko object name ke variable me rakhna

objects = [features for features in df1.columns if df1[features].dtypes=='O']
```

```python
objects
```

```python
# except classes kyunki classes numerical feature hai baki ko float me change karna

for i in objects:
    if i !='Classes':
        df1[i]=df1[i].astype(float)
```

### Explaratory data analysis 

```python
# cleaned data set me EDA

# day , month and year ko remove karna

df1_copy = df1.drop(['day' ,'month' ,'year'] , axis=1)
```

```python
df1_copy
```

```python
df1_copy.head()
```

```python
df1_copy.columns
```

```python
## classes wale column me cotegory 

df1_copy['Classes'].value_counts()
```

```python
## encoding of the categories in classes
# Classes me jyada categories hone par use sirf 2 category me devide karne ke liye 0,1 [ not fire = 0 , fire = 1]

df1_copy['Classes'] = np.where(df1_copy['Classes'].str.contains('not fire') , 0,1)
```

```python
df1_copy['Classes']
```

```python
df1_copy.head(5)
```

```python
df1_copy.tail(5)
```

```python
df1_copy['Classes'].value_counts()
```

```python
## plot density plot for all features

plt.style.use('seaborn')
df1_copy.hist(bins=50 , figsize=(20,15))
plt.show()
              
```

```python
## percentage for pie chart

percentage = df1_copy['Classes'].value_counts(normalize=True)*100
```

```python
percentage
```

```python
classlabels=["Fire","Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```
